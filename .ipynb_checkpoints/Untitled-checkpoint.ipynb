```python
from spring_mass import TweezedParticle
import numpy as np
import matplotlib.pyplot as plt
```

```python
system = TweezedParticle(75e-3, pressure=1e-5, omega=1119991.718)
```

```python
system._gamma_
```

```python
x = 0
v = 0
t = np.linspace(0, 1e6*1.40250e-08,10000 )
delta_t = np.diff(t)[0]
```

```python
xs = []
for i in t:
    xs.append(x)
    x, v = system.step(np.array([[x],[v]]), delta_t=delta_t)
```

```python
system._gamma_, system._m_, system._k_, delta_t
```

```python
plt.plot(xs)
```
